# Example Assignment

Let
$$
A = 
\begin{bmatrix} 
A_{0, 0} & \dots & A_{0, n - 1} \\
\vdots & \ddots & \vdots \\
A_{n - 1, 0} & \dots & A_{n - 1, n - 1}
\end{bmatrix} 
\in \mathbb R^{n \times n}
$$ 
be a big square matrix with independent random entries. What do you think $A^T A$ will look like? Let's find out with numpy.

## Setup

First, we need to import some packages and figure out how to make a random matrix.

```python
import numpy as np
```

```python
def make_matrix(n, seed=42):
    np.random.seed(seed)
    return np.random.randn(n, n)

A = make_matrix(5)
print("Here's A:")
print(A)
print("And here's A^T A:")
print(A.T @ A)
```

Okay, nice.

## Visualization

I have a hard time reading numbers in an array, so let's plot our matrices using `matshow`. ([Read the docs](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.matshow.html).) A random matrix just looks like a bunch of noise.

```python
from matplotlib import pyplot as plt

dims = 50
plt.matshow(make_matrix(dims))
plt.show()
```

What about $A^T A$?

```python
A = make_matrix(dims)
plt.matshow(A.T @ A)
plt.show()
```

Interesting. Let's check the distribution of entries in this matrix with a histogram. (Read the docs on [seaborn histplot](https://seaborn.pydata.org/generated/seaborn.histplot.html).)

```python
import seaborn
import pandas as pd

entries = (A.T @ A).reshape(-1)
on_diag = np.eye(dims).reshape(-1)
df = pd.DataFrame({'value': entries, 'on_diag': on_diag})
seaborn.histplot(data=df, x='value', hue='on_diag')
plt.show()
```
